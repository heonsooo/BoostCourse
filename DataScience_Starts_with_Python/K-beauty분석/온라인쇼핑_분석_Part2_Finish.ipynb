# 4. 분석과 시각화를 위한 tidy data

```python
df = df_raw.melt(id_vars=['국가(대륙)별','상품군별','판매유형별'],value_vars=['2014 1/4', '2014 2/4', '2014 3/4',
       '2014 4/4', '2015 1/4', '2015 2/4', '2015 3/4', '2015 4/4', '2016 1/4',
       '2016 2/4', '2016 3/4', '2016 4/4', '2017 1/4', '2017 2/4', '2017 3/4',
       '2017 4/4', '2018 1/4', '2018 2/4', '2018 3/4', '2018 4/4', '2019 1/4',
       '2019 2/4', '2019 3/4', '2019 4/4'],var_name ='기간',value_name='백만원')
# 컬럼명 변경
df 
```

### 5.2 금액을 수치데이터로 표현하기 위해 데이터 타입 변경하기

```python
# - 문자를 결측치로 변경하고 float 타입으로 변경하기
df['백만원']=df['백만원'].replace('-',pd.np.nan).astype(float)
df.head()
```

### 5.3 필요없는 데이터 제거하기

```python
# 합계 데이터는 따로 구할 수 있기 때문에 전체 데이터에서 제거한다.
df = df[(df['국가(대륙)별']!='합계')&(df['상품군별']!='합계')].copy()
df
```

```python
df.info()
```

```python
# 결측치 보기
df.isnull().sum()
```

# 6. K-Beauty 시각화

## 6.1 전체 상품군 판매액

```python
# 판매유형별 데이터는 일부 기간에는 '계'만 존재하기 때문에 
# 판매유형별 =='계' 데이터만 가져와야 평균 값을 구하는 그래프에서 올바른 값을 표현 할 수 있다.
df_total = df[df['판매유형별']=='계'].copy()
df_total.head()
```

```python
# 연도, 판매액으로 lineplot 그리기
plt.figure(figsize=(16,7))
sns.lineplot(data=df_total,x='연도', y='백만원')
```

```python
# 연도, 판매액으로 lineplot 그리고 상푼군별로 다른 색상으로 표시하기
plt.figure(figsize=(10,7))
sns.lineplot(data=df_total,x='연도', y='백만원', hue='상품군별')

plt.legend(bbox_to_anchor=(1.05,1), loc =2,borderaxespad=0.)
```

```python
# 위에 그린 그래프를 자세히 보기 위해 서프플롯으로 표시하기
sns.relplot(data=df_total, x='연도',y='백만원',
            hue='상품군별',kind='line',
           col='상품군별', col_wrap=4)
```

```python
# isin을 사용해 화장품,의류 및 패션 관련 상품 제외를 하고 df_sub이라는 변수에 담는다.
df_sub = df_total[~df_total['상품군별'].isin(['화장품','의류 및 패션 관련상품'])].copy()
df_sub
```

```python
# 연도별 판매액을 상품군별로 relplot을 활용해 서브플롯 그리기

sns.relplot(data=df_sub, x='연도',y='백만원',
            hue='상품군별',kind='line',
           col='상품군별', col_wrap=4)
```

## 6.2 화장품의 온라인쇼핑 해외 판매액

```python
# df_cosmetic 이라는 변수에 상품군별이 화장품인 데이터만 가져오기

df_cosmetic = df_total[df_total['상품군별'] =='화장품'].copy()
df_cosmetic['상품군별'].unique()
```

```python
# 연도와 판매액을 lineplot으로 그리고 분기별로 다른 색상으로 표현해보기.
plt.figure(figsize=(10,7))
sns.lineplot(data=df_cosmetic, x='연도', y='백만원',hue='분기',ci=False)
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터 시각화 하기
plt.figure(figsize=(15,5))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic, x='기간', y= '백만원')
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터 시각화 하고 '국가(대륙)별'로 다른 색상으로 표시하기

plt.figure(figsize=(15,5))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic, x='기간', y= '백만원',hue='국가(대륙)별')
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터 시각화 하고 중국을 제외한 '국가(대륙)별'로 다른 색상으로 표시하기

plt.figure(figsize=(15,5))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic[df_cosmetic['국가(대륙)별']!='중국'],
             x='기간', y= '백만원',hue='국가(대륙)별')
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터 시각화하고 '판매유형별'로 다른 색상으로 표시하기
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
df_sell = df[(df['판매유형별']!= '계') &( df['판매유형별'] != '면세점')].copy()

# 면세점 이외(온라인) 국가별 화장품 판매액
sns.lineplot(data=df_sell, x= '기간', y='백만원',hue = '국가(대륙)별', ci=None)
```

## 6.3 의류 및 패션 관련 상품 온라인 판매액

```python
# df_fashion 이라는 변수에 의류 데이터만 가져와 따로 담아주기

df_fashion = df[(df['상품군별'] =='의류 및 패션 관련상품') & (df['판매유형별']=='계')].copy()
df_fashion.head()
```

```python
# '의류 및 패션 관령상품'판매액에 대한 기간별 금액 데이터 시각화 하고 '국가(대륙)별'로 다른 색상으로 표시하기

plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion, x='기간', y='백만원',hue ='국가(대륙)별')
```

```python
# '의류 및 패션관련 상품' 판매액에 대한 기간별 금액 데이터 시각화하고 '판매유형별'로 다른 색상으로 표시하기
df_fashion2 = df[(df['상품군별']=='의류 및 패션 관련상품')& (df['판매유형별']!= '계')].copy()

plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion2,x='기간', y='백만원', hue ='판매유형별',ci=None)
```

## 6.4 데이터 집계하기

```python
# 피봇테이블로 '국가(대륙)별', '연도'별 합계 금액을 표 형태로 구하기
# df_fashion.pivot_table?
df_pivot= df_fashion.pivot_table(index='국가(대륙)별', columns='연도', values='백만원',aggfunc='sum')
df_pivot
```

# 7. 전체 상품군별로 온라인 쇼핑 해외 직접판매액은 증가했을까?

```python
# 연도별 판매액을 그리고 '상품군별'로 다른 색상으로 표현하기
plt.figure(figsize=(15,8))
sns.lineplot(data=df_total, x='연도',y='백만원',hue='상품군별')
plt.legend(bbox_to_anchor=(1,1), loc =2,borderaxespad=0.)
```

# 결론
- 2014년부터 2019년까지 해외직접판매가 해마다 증가했음을 알 수 있다.
- 특히 화장품 상품군의 비중이 매우크며, 국가로는 중국으로 가장 많은 수출을 한다.
- 따라서, 해외 직접 판매를 중국 국가를 타겟으로 화장품 상품군을 판매한다면, 시장성이 있을 것으로 판단 된다.
